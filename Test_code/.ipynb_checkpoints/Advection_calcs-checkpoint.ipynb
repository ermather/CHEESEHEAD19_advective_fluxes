## Hrz Advection Wrapping Function

```python
#Hrz advection calcs
'''
Function that caculates horizontal advection of H and LE

Parameters:
- hrz_grad_tow: horizontal gradients for given tower
- tow: tower
- step: step size for vertical profiles, defaule = 0.1 m

'''

def hrz_ad_calc(hrz_grad, dtindex, step = 0.1):
    
    H_adv = pd.DataFrame(index = dtindex, columns = towlist)
    LE_adv = pd.DataFrame(index = dtindex, columns = towlist)
    
    for tow in towlist:
        if tc.loc[tow]['z1'] == 30:
            #Eventually set this up to read in previously calculated data, rather than calculating each time
            #Especially for WS_porf #fixthis does that actually matter? It runs pretty fast...
            WS_prof = wind_prof(tow, step)
            grad_prof = prof_calcs(hrz_grad, tow, dtindex, step)  
            
            #H advection
            H_adv_step = grad_prof['TA'].values*WS_prof.mul(-rho[tow]*cp*step/2000, axis = 0) #Advection of sensible heat

            #LE advection
            LE_adv_step = grad_prof['H2O'].values*WS_prof.mul(-step*L/2000) #Advection of latend heat 


            # hrz_ad = (H_ad + LE_ad)
            H_adv[tow] = H_adv.sum(axis = 1)
            LE_adv[tow] = LE_adv.sum(axis = 1)
            #This outputs zeor if all nans, fiture out how to handle that #fixthis
    
    # return grad_prof, TA_K
    return H_adv, LE_adv
    
    #Deal with water vapor later
    # md_w = e/(R*TA_K) #Molar density of water vapor [mol/m^3], used in SLE calc
```

## Vert Advection Wrapping Function

```python
#Vert H and LE advection calcs
def vert_ad_calc(tow, step = 0.1):

    WS_prof = wind_prof(tow, step)
    
    #Calc vert T diff (only works for 30 m towers(only works for 30 m towers
    TA_30 = TA.loc[30][tow]
    TA_10 = TA.loc[10][tow]
    TA_2 = TA.loc[2][tow]
    TA_0 = TA_2 - (TA_10 - TA_2)*(2/8) #Surface temp
    TA_diff = TA_0 - TA_30 #Mistakenly had TA_10 instead of TA_0, fixed 11/1/2023
    
    #Vert [H2O] diff (only works for 30 m tows)
    H2O_30 = H2O.loc[30][tow]
    H2O_10 = H2O.loc[10][tow]
    H2O_2 = H2O.loc[2][tow]
    
    H2O_0 = H2O_2 - (H2O_10 - H2O_2)*(2/8) #Surface temp
    H2O_diff = H2O_0 - H2O_30 
    
    
    TA_K = TA.loc[30][tow] + CtoK #Change for shorter towers

    H_vert = TA_diff*vert_wind[tow]*rho[tow]*cp
    LE_vert= H2O_diff*vert_wind[tow]*L
    
    # H_vert = vert_wind[tower]
    # LE_vert= vert_wind[tower]*L
    
    
    return H_vert, LE_vert
    
    # md_w = e/(R*TA_K) #Molar density of water vapor [mol/m^3], used in SLE calc
```

## Profile Calcs

### Wind Profile

```python
#Function to calc wind speed at top of tower using ustar
def WS_top_calc(tow, d_rat = 0.65):
    k = 0.4 #
    
    #Measurement height (z1 for all towers except PFA)

    h_tow = tc.z1.loc[tow]
        
    h_veg = tc.veg_h.loc[tow] #Vegetation height
    WS_calc = (hrz_wind.loc['Ustar'][tow]/k)*np.log((h_tow - d_rat*h_veg)/(0.1*h_veg))
    return WS_calc



def wind_prof(tow, step = 0.1):
    towdat = tc.loc[tow] #Tower data
    
    #Step is step size to be used for integration
    k = 0.4 #Von Karmann constant
    h_veg = tc.veg_h.loc[tow] #vegetation height
    h_tow = tc.z1.loc[tow] #tower height
    z_range = np.arange(0, h_tow + step, step)
    z_m = 0.1*h_veg #momentum friction parameter
    z_m_soil = 0.006 #From biophysic notes, check book Table_, ch.5
    
    d = 0.65*h_veg #zero plane displacement
    
    # d_mean = h_tow - z_m*np.exp(k*WS_top[tower]/Ustar[tower])#zero plane displacement, mean of val calculated for all timestamps
    
    WS_prof = pd.DataFrame() #windspeed profile
    
    #reviewthis
    WS_calc = WS_top_calc(tow)
    WS_calc_rat = hrz_wind.loc['WS_top'][tow]/WS_calc
    u_of_h = hrz_wind.loc['Ustar'][tow]/k*np.log((h_veg-d)/z_m)*WS_calc_rat #wind speed at top of canopy
    u_of_soil = u_of_h*np.exp(0.1 - 1) #WS at 0.1h, using exponential 
    ustar_soil = u_of_soil*k/np.log(0.1*h_veg/z_m_soil) #Soil friction velocity, calculated from WS at 01.h
    #^This line used to just have h instead of h_veg, not 100% sure that h_veg is the right h
    #Look this up and come back to confirm later #fixthis see same issue below
    
    #Put in different func? #fixthis
    for z in z_range: 
        if z >= h_veg:
            WS_prof[z] = hrz_wind.loc['Ustar'][tow]/k*np.log((z-d)/z_m)*WS_calc_rat
        elif z>=0.1*h_veg:
            WS_prof[z] = u_of_h*np.exp(z/h_veg - 1) #fixthis - check if h_veg is correct h
        elif z>0:
            WS_prof[z] = ustar_soil/k*np.log((z)/z_m_soil)
        else:
            WS_prof[0] = 0
            
    return WS_prof
```

### TA and H2O gradient profiles

```python
def prof_calcs(hrz_grad, tow, dtindex, step = 0.1):
    towdat = tc.loc[tow] #Tower data
    
    h_tow = tc.z1.loc[tow] #tower height
    z_range = np.arange(0, h_tow + step, step)
    varlist = ['TA', 'H2O']
    
    hrz_grad_tow = hrz_grad.loc[dtindex[0]:dtindex[-1], (slice(None), slice(None), tow)]
    
    #Create dataframe to hold hrz gradient profile data
    hrz_grad_prof = make_multi_df([varlist, z_range], ["var", "z"], idx = dtindex) #Can get rid of start/enddate once I reload 

    # should maybe make this more flexible for different tower heights #fixthis
    # Vertical gradients of horizontal gradients
    gradgrad_up = (hrz_grad_tow.loc[:, (slice(None), 30)].values - hrz_grad_tow.loc[:, (slice(None), 10)].values)/20
    gradgrad_low = (hrz_grad_tow.loc[:, (slice(None), 10)].values - hrz_grad_tow.loc[:, (slice(None), 2)].values)/8    

    for z in z_range:     
        if towdat.z3 == 0: #Figure out how to handle different tower heights #fixthis
            pass
        else:
            
            if z >= 10: #towdat.z2:
                
                hrz_grad_prof.loc[:, (slice(None), z)] = hrz_grad_tow.loc[:, (slice(None), 10)].values + gradgrad_up*(z-towdat.z2)

            else:
                hrz_grad_prof.loc[:, (slice(None), z)] = hrz_grad_tow.loc[:, (slice(None), 10)].values - gradgrad_low*(towdat.z2-z) 
                
            
    return hrz_grad_prof

    
    
```

```python
err_df = pd.DataFrame(columns = ['gap_change_med', 'gap_change_mean', 'gap_change_med_cut', 'gap_change_mean_cut'])#,
                                # 'gap_change_hrz_ad_only_med', 'gap_change_hrz_ad_only_med'])
for tower in vert_ad.columns:
    filepath = tc.org_path.loc[tower]
    df = pd.read_csv(filepath)

    df = df.applymap(lambda x: np.nan if x == -9999 else x)

    dt_ts = pd.to_datetime(df['TIMESTAMP_START'], format = '%Y%m%d%H%M') 
    df.set_index(dt_ts, inplace = True)


    startdate = dt.datetime(2019, 6, 20, 1)
    enddate = dt.datetime(2019, 10, 14, 23)
    df = df[startdate:enddate]

    NETRAD = df.NETRAD_1_1_1
    LE_turb = df.LE_1_1_1
    H_turb = df.H_1_1_1
    G = df. G_1_1_1

    LE_ad_hrz = hrz_ad.loc['LE'][tower]
    H_ad_hrz = hrz_ad.loc['H'][tower]

    LE_ad_vert = vert_ad.loc['LE'][tower]
    H_ad_vert = vert_ad.loc['H'][tower]

    LE_ad_vert_cut = LE_ad_vert.where(LE_ad_vert>np.nanpercentile(LE_ad_vert, 10)).where(LE_ad_vert<np.nanpercentile(LE_ad_vert, 90))
    H_ad_vert_cut = H_ad_vert.where(H_ad_vert>np.nanpercentile(H_ad_vert, 10)).where(H_ad_vert<np.nanpercentile(H_ad_vert, 90))

    E_bal_org = NETRAD - LE_turb - H_turb - G

    E_bal_new = NETRAD - LE_turb - H_turb - G - LE_ad_hrz - H_ad_hrz - LE_ad_vert - H_ad_vert

    E_bal_new_cut = NETRAD - LE_turb - H_turb - G - LE_ad_hrz - H_ad_hrz - LE_ad_vert_cut - H_ad_vert_cut
    
    E_bal_hrz = NETRAD - LE_turb - H_turb - G - LE_ad_hrz - H_ad_hrz

    # E_out = LE_turb + H_turb + G + LE_ad_hrz + H_ad_hrz + LE_ad_vert + H_ad_vert

    # E_out_no_ad = LE_turb + H_turb + G


    err_org = E_bal_org/NETRAD
    err_new = E_bal_new/NETRAD
    err_new_cut = E_bal_new_cut/NETRAD
    err_hrz_ad_only = E_bal_hrz/NETRAD

    # plt.hist(err_new, bins = 50, density = True);
    # plt.hist(err_org, bins = 50, alpha = .5, density = True);

    err_change = np.abs(err_org) - np.abs(err_new)
    err_change_cut = np.abs(err_org) - np.abs(err_new_cut)
    err_change_hrz_only = np.abs(err_org) - np.abs(err_hrz_ad_only)
    # plt.hist(err_change, bins = 100)
    err_df.loc[tower] = [err_change.median(), err_change.mean(), err_change_cut.median(), err_change_cut.mean()]#,
                        # err_change_hrz_only.median(), err_change_hrz_only.mean()]

```
