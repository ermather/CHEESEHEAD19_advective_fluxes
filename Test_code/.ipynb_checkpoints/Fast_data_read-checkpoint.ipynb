```python
hours_f = np.arange(0,24) #Array of hours as floats
hours_s = [str(x).zfill(2) for x in hours_f] #Make list of hours as strings to use in calling files

w_vars = tc.wind_var

df_w = pd.DataFrame(index = dtindex, columns = tc.wind_var)


# df_w['ts_s'] =  df_w.index.strftime('%m%d_%H') #Add time stamp as string to use for reading in files
# ts_s =  df_w.index.strftime('%m%d_%H') #Add time stamp as string to use for reading in files
ts_s = pd.Series(data = df_w.index.strftime('%m%d_%H'), index = dtindex)
```

```python
for hour in dtindex:
    if (hour.hour == 1) and (hour.minute == 0):
        print(hour)

    url = 'http://co2.aos.wisc.edu/data/CHEESEHEAD-incoming/ISFS/high-rate/isfs_geo_hr_2019' +ts_s.loc[hour] + '.nc'
    with fsspec.open(url) as fobj:
        ds = xr.open_dataset(fobj)
            
    if hour.minute ==0:    
        for var in w_vars[1:]: #Skip WLEF - not in deataset
            df_w.loc[hour, var] = ds[var].mean(axis = 1).values[0:1800].mean(); 
            #first mean is second average (over 20 measurements/sec), second mean is 30 min avg
            
    elif hour.minute ==30:
        for var in w_vars[1:]:
            df_w.loc[hour, var] = ds[var].mean(axis = 1).values[1800:3600].mean();
            
```

```python
# Code to rename vert_wind columns - from var names to tow names
tow_names =  dict(zip(tc.wind_var, tc.index)) #Dictionary to relate 
df_w = df_w.rename(columns = tow_names)
```

```python
#Shift time from UTC to CST to match ameriflux data
index_CDT = df_w.index - dt.timedelta(hours = 6)
df_w_CST = df_w.set_index(index_CDT)
```
