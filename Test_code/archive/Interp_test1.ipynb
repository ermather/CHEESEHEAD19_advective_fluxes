```python
tx = [0, 5, 1, 10, 8] #x coordinates of towers
ty = [9, 5, 2, 1, 7] #y coordinates of towers

T_t = [3, 5, 7, 2, 6] #Tower temp

x = np.linspace(0, 10, 11)
y = np.linspace(0, 10, 11)
X, Y = np.meshgrid(x, y)

T_i = griddata((tx, ty), T_t, (X, Y), method = 'linear') #intermpolated temp
```

```python
plt.pcolormesh(T_i)
```

```python
Xf= X.flatten()
Yf= Y.flatten()

rbf3 = Rbf(tx, ty, T_t, function="multiquadric", smooth=5)
T_i2 = Rbf(Xf, Yf)

plt.pcolormesh(Xf, Yf, T_i2)
```
