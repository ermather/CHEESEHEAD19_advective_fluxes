```python
dtindex = readdata('dtindex')
```

```python
TA_2 = pd.DataFrame(index = dtindex, columns = tc.index) 
H2O_2 = pd.DataFrame(index = dtindex, columns = tc.index)

#Set vertical gradients to calc 10 and 30 m values
#Could make these functions of time
TA_vert_grad = 0 
H2O_vert_grad = 0

for tow in TA_2.columns:
    TA_2[tow] = 20 + 0.4 * tc['x'].loc[tow]
    H2O_2[tow] = 10 + 0.2 * tc['x'].loc[tow]
    
TA_10 = TA_2 + 8*TA_vert_grad
TA_30 = TA_2 + 28*TA_vert_grad
        
        
H2O_10 = H2O_2 + 8*H2O_vert_grad
H2O_30 = H2O_2 + 28*H2O_vert_grad
        
#Make same format at actual data, if I move height to column, need to do that here too
TA_all = pd.concat([TA_2, TA_10, TA_30], keys = [2, 10, 30])
H2O_all = pd.concat([H2O_2, H2O_10, H2O_30], keys = [2, 10, 30])

TA_all.to_pickle('../UnitTest/Inputs/TA_orig_all.pickle')
H2O_all.to_pickle('../UnitTest/Inputs/H2O_orig_all.pickle')
```

```python
WS = pd.DataFrame(index = dtindex, columns = tc.index)
WD = pd.DataFrame(index = dtindex, columns = tc.index)
Ustar = pd.DataFrame(index = dtindex, columns = tc.index)
WS[:] = 2
Ustar[:] = 0.4

for tow in tc.index:
    WD[tow] = np.arange(len(dtindex))%360

WD = WD.astype(float)
hrz_wind = pd.concat([WS, WD, Ustar], keys = ['WS_top', 'WD', 'Ustar'])
hrz_wind.to_pickle('../UnitTest/Inputs/wind_dat.pickle')
```
